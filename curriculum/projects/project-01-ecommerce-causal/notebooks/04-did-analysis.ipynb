# 04 双重差分分析（DID）

## 学习目标

- 理解DID的基本原理
- 掌握平行趋势检验
- 估计处理效应

## 背景

DID利用面板数据的结构，通过差分再差分消除时间不变的选择偏差。

---

## 步骤 1：数据准备

构造处理前和处理后的面板数据。

```python
# TODO: 加载数据并构造面板数据
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

df = pd.read_csv('../data/raw/ecommerce_618.csv')

# 构造长格式面板数据
df_pre = df[['user_id', 'treatment', 'pre_gmv']].copy()
df_pre['post'] = 0
df_pre['gmv'] = df_pre['pre_gmv']

df_post = df[['user_id', 'treatment', 'post_gmv']].copy()
df_post['post'] = 1
df_post['gmv'] = df_post['post_gmv']

panel = pd.concat([df_pre[['user_id', 'treatment', 'post', 'gmv']],
                   df_post[['user_id', 'treatment', 'post', 'gmv']]])

print(panel.head(10))
```

## 步骤 2：平行趋势检验

检验处理组和对照组在处理前是否有平行趋势。

```python
# TODO: 平行趋势检验
# 如果有处理前多期数据，可以画趋势图
# 这里简化：比较处理前两组的GMV差异

pre_diff = df[df['treatment']==1]['pre_gmv'].mean() - df[df['treatment']==0]['pre_gmv'].mean()
print(f'处理前差异: {pre_diff:.2f}')
print('如果差异不大，平行趋势假设更可信')
```

## 步骤 3：DID估计

使用回归模型估计DID系数。

```python
# TODO: DID回归
# Y = beta0 + beta1*treatment + beta2*post + beta3*(treatment*post) + epsilon
# beta3 就是DID估计量

model = smf.ols('gmv ~ treatment * post', data=panel).fit()
print(model.summary())

did_estimate = model.params['treatment:post']
print(f'\nDID估计量: {did_estimate:.2f}')
```
