## 七、方法论反思

### 7.1 方法选择反思

```
为什么用DML + Meta-Learners + 因果森林？

1. 不用简单回归 → 因为存在选择偏差
2. 不用AB实验 → 因为不能随机定价
3. 不用PSM → 因为高维匹配困难
4. 用DML → 因为需要处理高维协变量
5. 用Meta-Learners → 因为需要对比不同策略
6. 用因果森林 → 因为需要自动发现异质性
7. 用双重稳健 → 因为模型可能误设
```

### 7.2 迭代过程反思

```
第一次分析：简单回归 → 结论：加价有害 ❌
发现问题：忽略了选择偏差

第二次分析：加入控制变量 → 结论：效应减弱但仍为负 ⚠️
发现问题：控制变量不够，仍有遗漏

第三次分析：DML → 结论：效应为正 ✅
验证：通过稳健性检验

最终结论：在有效区域加价，在无效区域不加价
```

### 7.3 业务洞察反思

```
从"平均效应"到"异质性效应"：
- 平均效应告诉我们"总体上"怎么样
- 异质性效应告诉我们"在哪里"有效
- 业务需要的是后者

从"相关"到"因果"：
- 相关告诉我们"什么和什么一起变"
- 因果告诉我们"改变A会不会改变B"
- 定价策略需要的是后者
```
