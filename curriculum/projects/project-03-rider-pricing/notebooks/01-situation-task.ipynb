## 一、业务背景（Situation）

### 1.1 平台现状

某头部外卖平台面临骑手供需失衡问题：

| 场景 | 问题 | 影响 |
|------|------|------|
| 雨天/高峰 | 订单激增，骑手不足 | 准时率下降，用户投诉 |
| 偏远区域 | 骑手不愿接单 | 订单积压，用户体验差 |
| 平峰时段 | 骑手闲置 | 运力浪费，成本上升 |

### 1.2 各方立场

```
运营团队："加价能吸引更多骑手，提升准时率"
财务团队："加价会吓跑用户，订单量下降"
CEO："我要每个区域×每个时段的最优价格"
```

### 1.3 核心矛盾

这不是一个简单的"加不加"问题，而是：
- **在哪里加**（区域异质性）
- **什么时候加**（时段异质性）
- **加多少**（剂量效应）
- **加了对谁有效**（子群异质性）

**本质**：定价效应是高度异质的——雨天加价vs晴天加价，效应完全不同。

## 二、问题定义（Task）

### 2.1 分析目标

1. **评估动态定价对核心指标的影响**
   - 订单量（用户侧）
   - 准时率（体验侧）
   - 骑手供给（运力侧）

2. **识别异质性效应**
   - 哪些区域×时段加价有效？
   - 哪些区域×时段加价无效甚至有害？

3. **给出定价策略建议**
   - 差异化定价规则
   - ROI预期与风险提示

### 2.2 关键挑战

```
挑战1：不能随机定价 → 无法做理想AB实验
挑战2：效应高度异质 → 需要个性化估计
挑战3：模型可能误设 → 价格效应非线性
挑战4：选择偏差 → 雨天本来订单就少
```

## 四、业务干扰模拟

### 干扰1：运营团队的"行业惯例"

> "雨天必须加价30%，这是行业惯例！"

**你的思考**：
- 行业惯例 = 数据支撑吗？
- 30%是拍脑袋还是实验出来的？
- 不同区域、不同时段，30%都一样吗？

**后续发现**：数据显示加价20%和30%效果一样——边际效应递减。

### 干扰2：财务团队的"一刀切"

> "所有区域都必须有正ROI！"

**你的思考**：
- ROI的定义是什么？短期还是长期？
- 偏远区域和CBD能一样吗？
- 如果某个区域加价反而降低GMV，还要坚持正ROI吗？

**后续发现**：偏远区域加价反而降低总GMV——用户流失效应 > 骑手供给效应。

### 干扰3：CEO的时间压力

> "1周内出结论！"

**你的思考**：
- 1周能做什么？探索性分析 + 初步方向
- 完整结论需要多久？模型训练+验证需要2周
- 如何管理预期？区分"方向性结论"和"精确估计"

**应对策略**：
1. 第1周：快速探索，给出初步方向
2. 明确告知：精确估计需要2周
3. 阶段性汇报：让CEO看到进展
