### 参考答案

<details>
<summary>点击展开参考答案</summary>

**主要指标（OEC）**：停留时长（dwell_time）

**选择理由**：
1. **业务相关性**：停留时长是内容APP的核心北极星指标，直接反映用户参与度
2. **敏感性**：停留时长对新推荐算法的变化敏感，容易检测出效果
3. **可解释性**：停留时长提升意味着用户更满意推荐内容

**为什么不选其他指标**：
- CTR：可能受标题党影响，不代表真实满意度
- 次日留存：是长期指标，实验周期内难以观测显著变化
- DAU：受太多外部因素影响，难以归因于推荐算法

**Guardrail指标**：
1. **CTR（点击率）**：反映内容匹配质量。如果停留时长提升但CTR下降，说明推荐的内容吸引人但用户不主动点击，可能存在"标题党"问题
2. **次日留存**：反映用户长期满意度。如果短期停留时长提升但留存下降，说明新算法在"透支"用户信任

</details>
